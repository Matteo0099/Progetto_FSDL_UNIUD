```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

euro_intro = {
    'EURO 1': 1992,
    'EURO 2': 1995,
    'EURO 3': 1999,
    'EURO 4': 2005,
    'EURO 5': 2008,
    'EURO 6': 2014
}

def carica_dato(path):
    return pd.read_csv(path, sep=';', decimal=',')

pm25 = carica_dato('dataset/altrids/pm25.csv')
pm10 = carica_dato('dataset/altrids/pm10.csv')
nox  = carica_dato('dataset/altrids/nox.csv')

df = pd.DataFrame({
    'Anno': pm25['Category'].astype(int),
    'PM2.5': pm25['Trasporti Stradali'],
    'PM10': pm10['Trasporti Stradali'],
    'NOx': nox['NOx']
})
df_melted = df.melt(id_vars='Anno', var_name='Inquinante', value_name='Valore')

sns.set(style="whitegrid")
plt.figure(figsize=(16, 6))
sns.barplot(data=df_melted, x='Anno', y='Valore', hue='Inquinante')
anni = df['Anno'].tolist()
for label, year in euro_intro.items():
    if year in anni:
        idx = anni.index(year)
        plt.axvline(x=idx-0.94, color='gray', linestyle='--', alpha=0.8)
        plt.text(idx, plt.ylim()[1]*0.97, label, rotation=0,
                 color='gray', va='top', ha='center', fontsize=13, alpha=1)
        
#plt.title("PM2.5, PM10, NOx e Classi Euro")
plt.xlabel("Anni 90' - 2022")
plt.ylabel("Emissioni kt/a")
plt.legend(title='Inquinante')
plt.tight_layout()
plt.savefig('emissioni_veicoli_euro.png', dpi=350)
plt.show()
```
